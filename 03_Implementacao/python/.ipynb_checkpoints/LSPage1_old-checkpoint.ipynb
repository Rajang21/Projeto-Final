```python
import os
import voila
import ipywidgets
import import_ipynb
import fitz
import functools

#from LocalSearch import LocalSearch
from wooshTest import WhooshHandler
from wooshTest import Searcher
from ipywidgets import Button, Layout

from IPython.display import IFrame

titleLabel = ""
out = ipywidgets.Output(layout={'border': '1px solid black'})
outResult = ipywidgets.Output(layout={'border': '1px solid black', 'width': 'auto'})

def firstWidget():
    
    configureBtn = ipywidgets.Button(description="Configure Search")
    searchBtn = ipywidgets.Button(description="Search")
    buttons = [configureBtn,searchBtn]
    
    configureBtn.on_click(functools.partial(defineindex, operation = 'configure', btn = buttons))
    searchBtn.on_click(functools.partial(defineindex, operation = 'search',  btn = buttons))
    
    with out:
        display(configureBtn, searchBtn)
    return

def defineindex(event, operation, btn=""):
    global titleLabel
    
    titleIndex = ipywidgets.Label('Insert index folder name:')
    pathBoxIndex = ipywidgets.Text(placeholder = 'index')

    enterBtn = ipywidgets.Button(description="Enter")
    
    title = ""
       
    if(operation == 'configure'):
        title = "Configure Search Menu"
        
        if(btn != ""):
            btn[0].layout.visibility = 'hidden'
            btn[1].layout.visibility = 'visible'
        
        enterBtn.on_click(functools.partial(createConfigureSearch, indexName = pathBoxIndex))
        
    elif(operation == 'search'):
        title = "Search Menu"
        
        if(btn != ""):
            btn[1].layout.visibility = 'hidden'
            btn[0].layout.visibility = 'visible'
        
        enterBtn.on_click(functools.partial(createSearch, indexName = pathBoxIndex))
        
    titleLabel = ipywidgets.HTML(value = f"<b><font-weight='bold'>{title}</b>")
    
    outResult.clear_output()
    with outResult:
        display(titleLabel, titleIndex, pathBoxIndex, enterBtn)
    
    return

#---------------------------------------------------------------------

def createConfigureSearch(event, indexName, WH = ''):
    global titleLabel

    if(WH == ''):
        WH = WhooshHandler(indexName.value)
       
    if(not os.path.exists(indexName.value)):
        titleIndex = ipywidgets.Label('There is no folder with this name: ' + indexName.value)
    
        createIndexBtn = ipywidgets.Button(description="Create index folder and schema", layout=Layout(width='30%'))
        createIndexBtn.on_click(functools.partial(createSchema, indexName = indexName, WH = WH))
        
        goBackBtn = ipywidgets.Button(description="Go back", layout=Layout(width='30%'))
        goBackBtn.on_click(functools.partial(defineindex, operation = "configure"))
        
        outResult.clear_output()
        with outResult:
            display(titleLabel, titleIndex, createIndexBtn, goBackBtn)
    
    else:
        titleIndex = ipywidgets.Label('Index Name: ' + indexName.value)
        
        insertFolderBtn = ipywidgets.Button(description="Insert Folder of files into the index", 
                                            layout=Layout(width='30%'))
        insertFolderBtn.on_click(functools.partial(insertExercises, WH = WH, indexName = indexName, isFolder = True))

        insertFileBtn = ipywidgets.Button(description="Insert one file into the index", 
                                          layout=Layout(width='30%'))
        insertFileBtn.on_click(functools.partial(insertExercises, WH = WH, indexName = indexName, isFolder = False))
        
        outResult.clear_output()
        with outResult:
            display(titleLabel, titleIndex, insertFolderBtn, insertFileBtn)
    
    return

def createSchema(event, indexName, WH):
    global titleLabel
    display(WH)
    display(indexName)
    result = WH.createSchema()
    
    if (result == None or result == False):
        result = 'Unfortunately it was not possible to create the schema!'
    else:
        result = 'The schema was created successfully!'
    
    resultLabel = ipywidgets.Label(result)
    
    returnBtn = ipywidgets.Button(description="Return to configure search menu", 
                                    layout=Layout(width='30%'))
    returnBtn.on_click(functools.partial(createConfigureSearch, indexName = indexName, WH = WH))
    
    outResult.clear_output()
    with outResult:
        display(titleLabel, resultLabel, returnBtn)
    
    return

def insertExercises(event, WH, indexName, isFolder):
    global titleLabel
    
    titleIndex = ipywidgets.Label('Index Name: ' + indexName.value)
    
    if isFolder:
        label = 'Enter the exercises folder directory path:'
    else:
        label = 'Enter the exercise directory path:'
    
    folderPathLabel = ipywidgets.Label(label)
    pathBox = ipywidgets.Text(placeholder = 'C:\\\ ...')
    
    insertBtn = ipywidgets.Button(description="Insert")
    insertBtn.on_click(functools.partial(insertExercisesResult, WH = WH, 
                                         indexName = indexName, isFolder = isFolder, path = pathBox))
    
    outResult.clear_output()
    with outResult:
        display(titleLabel, titleIndex, folderPathLabel, pathBox, insertBtn)
        
    return


def insertExercisesResult(event, WH, indexName, isFolder, path):
    global titleLabel
    
    titleIndex = ipywidgets.Label('Index Name: ' + indexName.value)
    
    WH.writeSchema(path.value,isFolder)
    
    result = 'Folder added to the index successfully!'        
    resultLabel = ipywidgets.Label(result)
    
    returnBtn = ipywidgets.Button(description="Return to configure search menu", 
                                    layout=Layout(width='30%'))
    returnBtn.on_click(functools.partial(createConfigureSearch, indexName = indexName, WH = WH))
    
    outResult.clear_output()
    with outResult:
        display(titleLabel, titleIndex, resultLabel, returnBtn)
    
    return



def createSearch(event, indexName, S=''):
    global titleLabel
    
    if(not os.path.exists(indexName.value)):
        alertLabel = ipywidgets.Label('There is no folder with this name: ' + indexName.value)
        alertLabel2 = ipywidgets.Label('You need to configure search and its content')
        
        outResult.clear_output()
        with outResult:
            display(titleLabel, alertLabel, alertLabel2)
    else:
        if(S == ''):
            S = Searcher(indexName.value)

        titleIndex = ipywidgets.Label('Index Name: ' + indexName.value)
        keywordsLabel = ipywidgets.Label('Enter keywords:')
        keywordsBox = ipywidgets.Text(placeholder = 'Array, random, ...')

        typeLabel = ipywidgets.Label('Define the type of documents')
        dropDownType = ipywidgets.Dropdown(
            options=['all','pdf', 'py', 'tex'],
            value='all',
            description='Type:',
            disabled=False,
        )
        
        dateLabel = ipywidgets.Label('Define the date of documents')
        dropDownDate = ipywidgets.Dropdown(
            options=['all time','this year', 'this month', 'this week'],
            value='all time',
            description='Date:',
            disabled=False,
        )
        
        box = ipywidgets.Checkbox(False, description='Sort by date')

        searchBtn = ipywidgets.Button(description="Search keywords")
        searchBtn.on_click(functools.partial(searchKeyword, indexName = indexName, S = S,
                                             keywords = keywordsBox, docType = dropDownType))

        outResult.clear_output()
        with outResult:
            display(titleLabel, titleIndex, keywordsLabel, keywordsBox, 
                    typeLabel, dropDownType, dateLabel, dropDownDate, box, searchBtn)
    
    return

def searchKeyword(event, indexName, S, keywords, docType):
    global titleLabel
    
    titleIndex = ipywidgets.Label('Index Name: ' + indexName.value)
    returnBtn = ipywidgets.Button(description="Return to search menu", 
                                    layout=Layout(width='30%'))
    returnBtn.on_click(functools.partial(createSearch, indexName = indexName, S = S))
    
    results = S.parser(keywords.value, docType.value)

    outResult.clear_output()
    i = 0
    with outResult:
        display(titleLabel, titleIndex, returnBtn)
        for result in results:
            
            path = result[0]
            pathLabel = ipywidgets.Label(str(i)+ ' - ' + path)
            i += 1
            
            if(result[1] == 'pdf'):
                image = showPdf(path)
                
                outImage = ipywidgets.Output(layout={'border': '1px solid black', 'visibility': 'hidden'})
                
                
                showImageBtn = ipywidgets.Button(description="Open PDF", 
                                layout=Layout(width='15'))
                showImageBtn.on_click(functools.partial(showImage, outImage = outImage, image = image))
                
                pdfHBox = ipywidgets.HBox([pathLabel, showImageBtn])               
                display(pdfHBox, outImage)
            else:
                display(pathLabel)
                
            display(ipywidgets.Label(''))
    return

def showPdf(path):
    doc = fitz.open(path)
    page = doc.loadPage(0)
    pix = page.getPixmap()
    png = pix.getPNGData()
    
    image = ipywidgets.Image(
                value=png,
                format='png',
                width=400,
                height='auto'
            )
    
    return image
    
def showImage(event, outImage, image):
    
    if(outImage.layout.visibility == 'visible'):
        outImage.layout.visibility = 'hidden'
        outImage.clear_output()
    else:
        outImage.layout.visibility = 'visible'
        with outImage: 
            display(image)
    
    
def hideShowImage():
    return

firstWidget()
display(out)
display(outResult)

```
